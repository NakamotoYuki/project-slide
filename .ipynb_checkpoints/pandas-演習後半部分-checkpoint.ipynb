```python
import pandas as pd
```

```python
path = "data/1hUSDJPY.csv"

df = pd.read_csv(path, dtype = {"open":"float","high":"float","low":"float","close":"float","volume":"float"},parse_dates=["time"])
```

### 1

```python
#
# 
# df
df["pct"] = df.close.pct_change()
len(df[df["pct"]>0])
```

### 2

```python
df["ma25"] = df["close"].rolling(25).mean()
```

### 3

```python
df["pct5"] = df["ma25"].pct_change(5)
```
