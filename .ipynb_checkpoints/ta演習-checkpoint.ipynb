```python
def get_ohlcv(exchange="binance-futures", symbol="BTC/USDT", timeframe="1m", limit=1000, times=2):
    """
    Parameters
    ---------------------------
    exchange: str
        "binance-futures", "ftx", etc
    crypto_type: str
        "future", etc
    symbol: str
        "BTC/USDT"
    timeframe: str
        "1m", '3m' "5m", "1h", "4h", "1d"
    limit: int
        how many calandles you get in one request
        max is depending on exchanges
    times: int
        how many times you request

    Return
    ---------------------------
    df_ohlcv: pandas.DataFrame
        columns=["time", "open", "high", "low", "close", "volume"]
    """
    period = {'1m': 1 * 60, "3m":3 * 60, '5m': 5 * 60, '30m': 5 * 30, '1h': 60 * 60, '4h': 60 * 60 * 4, '1d': 24 * 60 * 60}
    resolution = period[timeframe]
    if exchange == "binance-futures":
        ex = ccxt.binance({'options': {'defaultType': "future"}})

    data = []
    start_time = int(time.time())*1000  - limit*resolution*1000
    for i in range(times):
        ohlcv = ex.fetch_ohlcv(symbol, timeframe=timeframe, since=start_time, limit=limit)
        df = pd.DataFrame(ohlcv, columns=["time", "open", "high", "low", "close", "volume"])
        if len(df) < 1:
            print("break")
            break
        
        data.append(df.iloc[::-1])
        start_time = start_time - limit*resolution*1000 # 時間は戻っていく
        time.sleep(0.1)
    df_ohlcv = pd.concat(data)
    df_ohlcv = df_ohlcv.iloc[::-1]
    return df_ohlcv.reset_index(drop=True)
```

```python
# データ取得
df = get_ohlcv(exchange="binance-futures", symbol="BTC/USDT", timeframe="1m", limit=1000, times=2)

```

```python
# 課題１
# ボリンジャーバンドをtaで作成してみましょう

# 課題２
# ボリンジャーバンドを自分で作成してみましょう

# 課題３
# 関数にしてみましょう
```

```python
# 課題1

from ta.volatility import BollingerBands

# Initialize Bollinger Bands Indicator
indicator_bb = BollingerBands(close=df["close"], window=20, window_dev=2)

# Add Bollinger Bands features
df['bb_bbm'] = indicator_bb.bollinger_mavg()
df['bb_bbh'] = indicator_bb.bollinger_hband()
df['bb_bbl'] = indicator_bb.bollinger_lband()

# Add Bollinger Band high indicator
df['bb_bbhi'] = indicator_bb.bollinger_hband_indicator()
```

```python
df
```

```python
# 課題２

from ta.trend import SMAIndicator

MA2 = SMAIndicator(close=df["close"], window=20, fillna=True).sma_indicator()
std = MA2.std()
df["bb_bbh2"] = MA2 + std * 2
df["bb_bbl2"] = MA2 - std * 2
```

```python
# 課題3
```

```python
def bolinger_calc(data, term, sigma1, sigma2):
    MA2 = SMAIndicator(close=data["close"], window=term, fillna=True).sma_indicator()
    std = MA2.std()
    df["bb_bbh2"] = MA2 + std * sigma1
    df["bb_bbl2"] = MA2 - std * sigma2
    
    return df
```

```python
bolinger_calc(df, 20, 2, 2)
```
