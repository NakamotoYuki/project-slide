### 演習4

```python
path = "data/1hUSDJPY.csv"
df = pd.read_csv(path, parse_dates=["time"])
# [1]ボディ部分の作り方
# 1.open-close
# 2.絶対値にする方法→df[カラム名].abs()
# [2]ロウソク足の大きさ　high - low
# [3] [1]と[2]で作ったものを割り算する
```

```python
df["body"] = df["open"] - df["close"]
```

```python
df["body"] = df["body"].abs()
```

```python
df["candle_size"] = df.high - df.low
```

```python
df["ratio"] = df["body"]/df["candle_size"]
```
